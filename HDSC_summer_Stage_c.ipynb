```python
# Define URL
url = 'https://www.lineups.com/articles/top-10-best-soccer-players-in-the-world/'
# Ask hosting server to fetch url
requests.get(url)
```

```python
soup = BeautifulSoup(requests.get(url).content)
```

```python
print(soup.prettify())
```

```python
table = soup.select('table')[0]
columns = table.find('thead').find_all('th')
columns
```

```python
table_df = pd.read_html(str(table))[0]
table_df
```

```python
# Export to csv
table_df.to_csv('top_football_players.csv', index=False)
# Try to read csv
Top_10_football_players = pd.read_csv('top_football_players.csv')
```
