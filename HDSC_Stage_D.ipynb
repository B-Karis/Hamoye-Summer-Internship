```python
#Import all necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
from datetime import date
```

```python
# Read CSV File
Food_supply = pd.read_csv("c:\\users\\blessing.onumaegbu\\Downloads\\Africa Food Supply (2004 - 2013).csv", encoding="latin-1")
Food_production = pd.read_csv("c:\\users\\blessing.onumaegbu\\Downloads\\Africa Food Production (2004 - 2013).csv", encoding="latin-1")
```

```python
#converting year column to datetime
pd.to_datetime(Food_production['Year'])
```

```python
#grouping the data for easy analysis
Food_production.groupby(['Country', 'Year', 'Item']).sum()
Food_production.groupby(['Country', 'Year'])['Value'].sum()
Food_production = Food_production.drop('Item',axis=1)
Food_production.head()
```

```python
#Visualizing the data using seaborn
Food_production = sns.lineplot(x=Food_production["Year"], y=Food_production["Value"], hue=Food_production["Country"], palette = 'bright', ci = None, data=Food_production)
Food_production.set_yscale('log')
Food_production.legend(loc='upper right', bbox_to_anchor=(1.7, 1))
```

```python
##converting year column to datetime
pd.to_datetime(Food_supply['Year'])
```

```python
#grouping the data for easy analysis
Food_supply = Food_supply.groupby(['Country', 'Year']).sum().reset_index()
Food_supply.head()
```

```python
##Visualizing the data using seaborn
Food_supply = sns.lineplot(x=Food_supply["Year"], y=Food_supply["Value"], hue=Food_supply["Country"], palette = 'bright', ci = None, data=Food_supply)
Food_supply.set_yscale('log')
Food_supply.legend(loc='upper right', bbox_to_anchor=(1.7, 1))
```
